### Q1

```python
def wind_condition(speed):
    if speed < 5:
        return "Calm"
    elif 5 <= speed <= 20:
        return "Moderate"
    else:
        return "Strong"

print(wind_condition(3))  
print(wind_condition(10)) 
print(wind_condition(25)) 

'''
Giải thích chi tiết:
Hàm wind_condition(speed):

Hàm này nhận vào một tham số speed, biểu thị tốc độ của gió.
Các điều kiện trong hàm:

if speed < 5: Nếu tốc độ gió nhỏ hơn 5, hàm sẽ trả về chuỗi "Calm", nghĩa là "gió yên tĩnh".
elif 5 <= speed <= 20: Nếu tốc độ gió trong khoảng từ 5 đến 20 (bao gồm cả 5 và 20), hàm sẽ trả về "Moderate", nghĩa là "gió vừa phải".
else: Nếu tốc độ gió lớn hơn 20, hàm sẽ trả về "Strong", nghĩa là "gió mạnh".
Ví dụ sử dụng hàm:

print(wind_condition(3)): Gọi hàm với speed = 3. Do 3 < 5, hàm sẽ trả về "Calm".
print(wind_condition(10)): Gọi hàm với speed = 10. Do 5 <= 10 <= 20, hàm sẽ trả về "Moderate".
print(wind_condition(25)): Gọi hàm với speed = 25. Do 25 > 20, hàm sẽ trả về "Strong".
'''
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming data is loaded into a DataFrame, e.g., `precipitation_data`
# precipitation_data = pd.read_csv('precipitation_data.csv')
# Here, I’ll create a sample dataset to work with for illustration.

# Sample data (Replace with actual data from your file)
precipitation_data = pd.DataFrame({
    'Day': range(1, 31),
    'Precipitation': [0, 2, 0, 7, 0, 0, 5, 0, 3, 8, 0, 0, 10, 1, 0, 0, 4, 6, 0, 0, 0, 5, 2, 9, 0, 0, 11, 3, 0, 0]
})

# Bar Plot for daily precipitation levels
plt.figure(figsize=(10, 6))
plt.bar(precipitation_data['Day'], precipitation_data['Precipitation'], color='blue')
plt.xlabel('Day')
plt.ylabel('Precipitation (mm)')
plt.title('Daily Precipitation Levels')
plt.show()

# Pie chart for rain categories
no_rain = (precipitation_data['Precipitation'] == 0).sum()
light_rain = (precipitation_data['Precipitation'] < 5).sum()
heavy_rain = (precipitation_data['Precipitation'] >= 5).sum()

plt.figure(figsize=(6, 6))
plt.pie([no_rain, light_rain, heavy_rain], labels=['No rain', 'Light rain', 'Heavy rain'], autopct='%1.1f%%')
plt.title('Rainfall Distribution')
plt.show()

'''
Thư viện được sử dụng:

matplotlib.pyplot: Được sử dụng để tạo các biểu đồ và đồ thị.
pandas: Được sử dụng để xử lý và phân tích dữ liệu, với dữ liệu được lưu trong DataFrame.
Dữ liệu giả định:

Đoạn code giả định rằng dữ liệu về lượng mưa hàng ngày được lưu trong một DataFrame có tên precipitation_data. Để minh họa, dữ liệu giả định được tạo thủ công, với hai cột:
'Day': Số thứ tự ngày (từ 1 đến 30).
'Precipitation': Lượng mưa trong ngày đó (đơn vị là mm). Các ngày có lượng mưa bằng 0 nghĩa là không có mưa.
Biểu đồ cột (Bar Plot) cho lượng mưa hàng ngày:

plt.figure(figsize=(10, 6)): Tạo một khung hình với kích thước 10x6 inches.
plt.bar(precipitation_data['Day'], precipitation_data['Precipitation'], color='blue'): Tạo biểu đồ cột với trục x là các ngày và trục y là lượng mưa. Cột màu xanh đại diện cho mức lượng mưa.
plt.xlabel('Day'): Đặt nhãn cho trục x là "Day" (Ngày).
plt.ylabel('Precipitation (mm)'): Đặt nhãn cho trục y là "Precipitation (mm)" (Lượng mưa, tính bằng mm).
plt.title('Daily Precipitation Levels'): Đặt tiêu đề cho biểu đồ là "Daily Precipitation Levels".
plt.show(): Hiển thị biểu đồ cột.
Biểu đồ tròn (Pie Chart) cho các loại lượng mưa:

Tính toán các danh mục lượng mưa dựa trên giá trị lượng mưa hàng ngày:
no_rain: Số ngày không có mưa (Precipitation == 0).
light_rain: Số ngày có lượng mưa dưới 5mm (Precipitation < 5).
heavy_rain: Số ngày có lượng mưa từ 5mm trở lên (Precipitation >= 5).
plt.figure(figsize=(6, 6)): Tạo một khung hình kích thước 6x6 inches cho biểu đồ tròn.
plt.pie([no_rain, light_rain, heavy_rain], labels=['No rain', 'Light rain', 'Heavy rain'], autopct='%1.1f%%'): Tạo biểu đồ tròn với 3 phần:
no_rain, light_rain, và heavy_rain, tương ứng với các loại lượng mưa không mưa, mưa nhẹ, và mưa nhiều.
labels=['No rain', 'Light rain', 'Heavy rain']: Đặt nhãn cho từng phần của biểu đồ tròn.
autopct='%1.1f%%': Hiển thị phần trăm của từng loại trong biểu đồ.
plt.title('Rainfall Distribution'): Đặt tiêu đề cho biểu đồ tròn là "Rainfall Distribution".
plt.show(): Hiển thị biểu đồ tròn.
'''
```

```python
import random

for hour in range(1, 13):
    temperature = random.uniform(-5, 10)  # Example temperatures
    try:
        if temperature < 0:
            raise ValueError(f"Temperature dropped below 0°C at hour {hour}. Stopping recordings.")
        print(f"Hour {hour}: {temperature:.2f}°C")
    except ValueError as e:
        print(e)
        break
```

```python
# Constants
c1 = -8.78469475556
c2 = 1.61139411
c3 = 2.33854883889
c4 = -0.14611605

# Function to calculate heat index
def calculate_heat_index(T, H):
    HI = c1 + (c2 * T) + (c3 * H) + (c4 * T * H)
    return HI

# Example calculation
T = 30
H = 70
print(f"Heat Index: {calculate_heat_index(T, H):.2f}°C")
```
